# 다변량 데이터로 다변량 그래프 그리기 


```python
import seaborn as sns 
import pandas as pd
anscombe = sns.load_dataset("anscombe")

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0 
    else:
        return 1
tips = sns.load_dataset("tips") 
print(tips)
```

```python
## 각원소별로 색을 지정하겠다.
## 그 내용들을 새로운 항으로 추가
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
scatter_plot = plt.figure() 
axes1 = scatter_plot.add_subplot(1, 1, 1) 
axes1.scatter(
    x=tips['total_bill'],  #x 데이터 
    y=tips['tip'],         #y 데이터 
    s=tips['size'] * 10,    
    c=tips['sex_color'],
    alpha=0.5)

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size') 
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip')
```

# 1. seaborn 라이브러리로 산점도 그래프 그리기¶


```python
ax = plt.subplots() 
ax = sns.regplot(x='total_bill', y='tip', data=tips) ##데이터 중에 편차가 제일 작은 값을 찾아준다. 
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots() 
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False) 
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips) # 위에 그래프가 또 생성됨 
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex") ##육각형으로 그려짐 
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
ax = plt.subplots() 
ax = sns.kdeplot(data=tips['total_bill'],
                 data2=tips['tip'], 
                 shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

# 데이터 합치기

```python
import pandas as pd

df1 = pd.read_csv('./data/concat_1.csv') 
df2 = pd.read_csv('./data/concat_2.csv') 
df3 = pd.read_csv('./data/concat_3.csv')
print(df1,'\n', df2,'\n', df3,'\n',)
```

```python
print(df1,'\n', df2,'\n', df3,'\n',)
row_concat = pd.concat([df1, df2, df3])  ## 데이터를 합치는건 []안에 넣어야 한다. 
print(row_concat)
print(row_concat.iloc[2])
```

```python
## 컬럼이 같으면 같은 커럼에 넣는다 
new_row_df = pd.DataFrame([['n1', 'n2', 'n3', 'n4']], 
                          columns=['A', 'B', 'C', 'D']) 
print(new_row_df)
```

```python
print(pd.concat([df1,new_row_df],ignore_index=True))
```

```python
print(df1.append(new_row_df))
```

```python
## 행으로 붙음 ## 컬럼 이름이 반복됨 ABCDABCD
import pandas as pd 
col_concat1 = pd.concat([df1, df2, df3], axis=1) 
print(col_concat1)

## 열로 붙음
col_concat2 = pd.concat([df1, df2, df3], axis=0) 
print(col_concat2)

col_concat2.sum('B',axis=1)
```

```python
## Index 새로 부여 
## 행으로 붙음 ## 컬럼 이름이 반복됨 ABCDABCD
col_concat1 = pd.concat([df1, df2, df3], axis=1,ignore_index=True) 
print(col_concat1)

## 열로 붙음
col_concat2 = pd.concat([df1, df2, df3], axis=0,ignore_index=True) 
print(col_concat2)
```

# 공통 열과 공통 인덱스만 연결하기

# 누락값 개수 구하기

```python
ebola = pd.read_csv('./data/country_timeseries.csv')
```

```python
print(ebola.info())
print(ebola.count())
```

```python
# 전체 갯수 num_rows = ebola.shape[0]
print(num_rows)
```

```python
## 없는 데이터 개수 카운트
## 전체 갯수 - 각 항목당 개수
num_missing = num_rows - ebola.count() 
print(num_missing)
```

```python
#그런데 Numpy를 이용하면 바로 계산 할 수 있음 
import numpy as np 

print(np.count_nonzero(ebola.isnull()))
```

```python
# 각 항목에서 카운트 하는것
print(np.count_nonzero(ebola['Cases_Guinea'].isnull()))
```

# 누락값 처리하기 => .fillna 

```python
# 비교 할 기준값
print(ebola.iloc[0:10, 0:5])
```

```python
# 0 채워넣기
print(ebola.fillna(0).iloc[0:10, 0:5])
```

```python
# 위에 (=앞에) 값을 채워 넣어라! 
#카테고리 형은 안된다. 
print(ebola.fillna(method='ffill').iloc[0:10, 0:5])
```

```python
# 뒤에 값을 앞으로 채워줘라! 
print(ebola.fillna(method='bfill').iloc[0:10, 0:5])
```

```python
## .interpolate() => 보간법을 이용한다. 
## 선형 보간법
print(ebola.interpolate().iloc[0:10, 0:5])
```

```python
import urllib.request as ul 
import json, datetime, time
import pandas as pd 
movieDate = time.strftime('%Y%m%d', time.localtime(time.time()))

print(movieDate)
movieDate = 20200130
```

```python
url = f"http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key=66e652e1d2656b42f10d93c91e0295e4&targetDt={movieDate}"

request = ul.Request(url)
print(url)

response = ul.urlopen(request)
print(response)

rescode = response.getcode()
print(rescode)

if(rescode == 200) :
    responseDate = response.read()
```

```python
result = json.loads(responseDate)
print(result)
```

```python
pre_result = result['boxOfficeResult']
print(pre_result)
```

```python
pre_result1 = pre_result["dailyBoxOfficeList"]
print(pre_result1)
```

```python
cine = [{}]
for i in range(0,len(pre_result1)) :
    pre_result1[i]['targetDt'] =movieDate
    cine.append(pre_result1[i])
print(cine)
```

```python
datetime_obj = datetime.datetime.strptime(str(movieDate), '%Y%m%d').date()
datetime_obj_tmp = datetime_obj - datetime.timedelta(days=30)

day = datetime_obj_tmp.strftime("%Y-%m-%d").split('-')
movieDate = day[0]+day[1]+day[2]
print(movieDate)
```
